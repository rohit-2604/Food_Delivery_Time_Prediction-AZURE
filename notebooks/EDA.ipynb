## One Hot Encoding:

```python
def one_hot_encoder_func(data, column_name):
    """
    Perform one-hot encoding for a categorical column in a DataFrame.

    Parameters:
        data (DataFrame): The input DataFrame containing the data to be encoded.
        column_name (str): The name of the categorical column to be one-hot encoded.

    Returns:
        DataFrame: The input DataFrame with one-hot encoded columns added and the original column dropped.
    """
    # Validate inputs
    if not isinstance(data, pd.DataFrame):
        raise ValueError("Input 'data' must be a pandas DataFrame.")
    
    if column_name not in data.columns:
        raise ValueError(f"Column '{column_name}' not found in the DataFrame.")

    # Perform one-hot encoding
    one_hot_encoded = pd.get_dummies(data[column_name], prefix=column_name, dtype=int)

    # Concatenate the one-hot encoded columns with the original dataframe
    data = pd.concat([data, one_hot_encoded], axis=1)

    # Drop the original column
    data.drop(column_name, axis=1, inplace=True)

    return data
```

```python

def custom_pairbarplot(data, columns_to_compare, input_column_type, output_column,type_of_plot):

    # Grouping and summing each order type based on festival status
    grouped = data.groupby(output_column)[columns_to_compare].sum()

    # Plotting the bar plots
    ax = grouped.plot(kind=type_of_plot, figsize=(10, 4))
    ax.set_xlabel(f"{output_column} Status")
    ax.set_ylabel('Sum Count')
    ax.set_title(f'Comparison of {input_column_type} for {output_column} Status')

    plt.xticks(rotation=0)
    plt.legend(title=f"{output_column} Status")
    plt.show()
```
